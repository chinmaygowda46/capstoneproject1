```python
import os
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, auc
)
import joblib

warnings.filterwarnings("ignore")

FILE_PATH = "Dropout Dataset.csv"     # <-- update if your file name differs
OUTPUT_DIR = "dropout_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
RANDOM_STATE = 42

print("Working directory:", os.getcwd())
print("Dataset path:", FILE_PATH)
```

```python
df = pd.read_csv(FILE_PATH, encoding='utf-8')
print("\nShape:", df.shape)
display(df.head())

print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False))

# Detect target column
target_col = None
for col in df.columns:
    if col.lower() in ['target', 'dropout', 'status', 'result', 'outcome']:
        target_col = col
        break

if target_col is None:
    raise ValueError("❌ No target column found. Rename or set 'target_col' manually.")
else:
    print(f"\n✅ Target column detected: '{target_col}'")
    print(df[target_col].value_counts())
```

```python
plt.figure(figsize=(6, 4))
df[target_col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Target Distribution')
plt.xlabel(target_col)
plt.ylabel('Count')
plt.tight_layout()
plt.show()

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(num_cols) > 0:
    df[num_cols[:min(6, len(num_cols))]].hist(bins=20, figsize=(12, 6), color='lightblue', edgecolor='black')
    plt.tight_layout()
    plt.show()

if len(num_cols) > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[num_cols].corr(), annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
```

```python
le_target = LabelEncoder()
df[target_col + "_enc"] = le_target.fit_transform(df[target_col])
print("\nTarget classes:", dict(zip(le_target.classes_, le_target.transform(le_target.classes_))))

# ---- 2) Choose actual columns from your dataset ----
# Check your real dataset column names using:
# print(df.columns.tolist())
# Then, replace the placeholders below with the exact names found in your dataset.

selected = [
    'Curricular units 1st sem (grade)',          # academic performance
    'Curricular units 2nd sem (grade)',          # attendance or continued performance
    "Mother's qualification",                    # parental education
    "Father's qualification",                    # parental education
    'International'                              # geographic location (1 = foreign, 0 = local)
]

# ---- 3) Verify that all selected columns exist ----
missing = [col for col in selected if col not in df.columns]
if missing:
    raise ValueError(f"❌ Missing expected columns in dataset: {missing}")
else:
    print("✅ All selected columns found.")

# ---- 4) Assign feature matrix and target ----
X = df[selected]
y = df[target_col + "_enc"]

# ---- 5) One-hot encode categorical (object/string) columns ----
non_numeric = X.select_dtypes(include=['object', 'category']).columns.tolist()
print("\nCategorical columns to encode:", non_numeric)
if non_numeric:
    X = pd.get_dummies(X, columns=non_numeric, drop_first=True)

print("✅ Using selected features only:", X.columns.tolist())
print("✅ Final feature count:", X.shape[1])
```

```python
unique_classes = np.unique(y)
if len(unique_classes) > 1 and all(y.value_counts() > 1):
    stratify_param = y
else:
    stratify_param = None

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=stratify_param
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
joblib.dump(scaler, os.path.join(OUTPUT_DIR, "scaler.joblib"))

print("✅ Train/Test split complete. Train:", X_train.shape, "Test:", X_test.shape)
```

```python
models = {}

# Logistic Regression
print("\nTraining Logistic Regression...")
lr = LogisticRegression(max_iter=1500, solver='liblinear', random_state=RANDOM_STATE)
lr.fit(X_train_scaled, y_train)
models['LogisticRegression'] = {'model': lr, 'uses_scaled': True}
print("✅ Logistic Regression trained.")

# Random Forest
print("Training Random Forest...")
rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1)
rf.fit(X_train, y_train)
models['RandomForest'] = {'model': rf, 'uses_scaled': False}
print("✅ Random Forest trained.")

# XGBoost or fallback
try:
    from xgboost import XGBClassifier
    print("Training XGBoost...")
    xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss',
                        random_state=RANDOM_STATE, n_jobs=-1, verbosity=0)
    xgb.fit(X_train, y_train)
    models['XGBoost'] = {'model': xgb, 'uses_scaled': False}
    print("✅ XGBoost trained.")
except Exception as e:
    print(f"⚠️ XGBoost unavailable ({e}). Using fallback.")
    hgb = HistGradientBoostingClassifier(random_state=RANDOM_STATE)
    hgb.fit(X_train_scaled, y_train)
    models['XGBoost_fallback'] = {'model': hgb, 'uses_scaled': True}
    print("✅ HistGradientBoosting fallback trained.")

# Save feature names for future alignment
joblib.dump(X.columns.tolist(), os.path.join(OUTPUT_DIR, "training_columns.joblib"))
print("✅ Saved training feature list.")
```

```python
eval_list = []
binary = len(np.unique(y)) == 2

for name, info in models.items():
    model = info['model']
    X_eval = X_test_scaled if info['uses_scaled'] else X_test

    y_pred = model.predict(X_eval)
    acc = accuracy_score(y_test, y_pred)
    rep = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    cm = confusion_matrix(y_test, y_pred)
    f1 = rep['macro avg']['f1-score']

    roc_auc = None
    try:
        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_eval)
            roc_auc = roc_auc_score(y_test, y_proba[:, 1]) if binary else roc_auc_score(y_test, y_proba, multi_class='ovr')
    except:
        pass

    eval_list.append({
        'model': name, 'accuracy': acc, 'f1': f1, 'roc_auc': roc_auc,
        'confusion_matrix': cm, 'report': rep
    })
    print(f"\n=== {name} ===")
    print(f"Accuracy: {acc:.4f}, F1: {f1:.4f}, ROC AUC: {roc_auc}")
    print("Confusion matrix:\n", cm)

# Save evaluation summary
eval_df = pd.DataFrame([{'model': e['model'], 'accuracy': e['accuracy'], 'f1': e['f1'], 'roc_auc': e['roc_auc']} for e in eval_list])
eval_df.to_csv(os.path.join(OUTPUT_DIR, "evaluation_summary.csv"), index=False)
for name, info in models.items():
    joblib.dump(info['model'], os.path.join(OUTPUT_DIR, f"{name}_model.joblib"))
print("\n✅ Saved evaluation summary & models.")
```

```python
if 'RandomForest' in models:
    rf = models['RandomForest']['model']
    fi = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_}).sort_values('Importance', ascending=False)
    fi.to_csv(os.path.join(OUTPUT_DIR, "rf_feature_importances.csv"), index=False)
    print("\nTop RandomForest Features:\n", fi.head(10))
```

```python
%%writefile app.py
# =========================================================
# Student Dropout Prediction – Streamlit App
# =========================================================
import streamlit as st
import pandas as pd
import joblib
import numpy as np

# ---------------------------------------------------------
# Load model and scaler
# ---------------------------------------------------------
MODEL_PATH = "dropout_outputs/RandomForest_model.joblib"
SCALER_PATH = "dropout_outputs/scaler.joblib"

model = joblib.load(MODEL_PATH)
scaler = joblib.load(SCALER_PATH)

st.set_page_config(page_title="Student Dropout Predictor", page_icon="🎓", layout="centered")

st.title("🎓 Student Dropout Prediction System")
st.markdown("This app predicts the probability that a student may drop out based on key academic and family factors.")

# ---------------------------------------------------------
# Input form
# ---------------------------------------------------------
st.header("🧮 Enter Student Details")

col1, col2 = st.columns(2)
with col1:
    sem1 = st.number_input("Curricular units 1st sem (grade)", min_value=0.0, max_value=100.0, value=70.0)
    mother_qual = st.selectbox("Mother's qualification", ["Primary", "Secondary", "Graduate"])
    international = st.selectbox("International (0 = Local, 1 = International)", [0, 1])
with col2:
    sem2 = st.number_input("Curricular units 2nd sem (grade)", min_value=0.0, max_value=100.0, value=75.0)
    father_qual = st.selectbox("Father's qualification", ["Primary", "Secondary", "Graduate"])

# Prepare input dictionary
input_dict = {
    "Curricular units 1st sem (grade)": sem1,
    "Curricular units 2nd sem (grade)": sem2,
    "Mother's qualification": mother_qual,
    "Father's qualification": father_qual,
    "International": international
}

# ---------------------------------------------------------
# Prediction
# ---------------------------------------------------------
if st.button("🔍 Predict Dropout Probability"):
    X_new = pd.DataFrame([input_dict])

    # One-hot encode categorical columns like during training
    non_numeric = X_new.select_dtypes(include=['object', 'category']).columns.tolist()
    if non_numeric:
        X_new = pd.get_dummies(X_new, columns=non_numeric, drop_first=True)

    # Align to model features
    for col in model.feature_names_in_:
        if col not in X_new.columns:
            X_new[col] = 0
    X_new = X_new[model.feature_names_in_]

    # Scale numeric columns
    try:
        X_scaled = scaler.transform(X_new)
    except Exception:
        X_scaled = X_new

    # Predict
    y_pred = model.predict(X_scaled)
    prob = model.predict_proba(X_scaled)[:, 1][0] * 100
    status = "Graduate" if y_pred[0] == 0 else "Dropout"

    # Display results
    st.subheader("🎓 Prediction Result")
    st.write(f"**Predicted Status:** {status}")
    st.write(f"**Dropout Probability:** {prob:.2f}%")

    if prob > 60:
        st.error("⚠️ High Risk of Dropout — consider student support measures.")
    elif prob > 30:
        st.warning("⚠️ Moderate Risk of Dropout — needs academic/financial monitoring.")
    else:
        st.success("✅ Low Risk of Dropout — student likely to continue successfully.")

st.markdown("---")
st.caption("🎓 Developed with ❤️ for the Student Dropout Analysis Capstone Project using Random Forest Classifier.")
```
